```python
def booksFromIDs(bookinfo,IDlist):
  df=pd.DataFrame(columns=bookinfo.columns)
  for id in IDlist:
    book=bookinfo[bookinfo.book_id==id]
    df=pd.concat([df,book],axis=0)
  df.reset_index(inplace=True,drop=True)
  return df
```

```python
booksFromIDs(bookinfo,[1,2])
```
